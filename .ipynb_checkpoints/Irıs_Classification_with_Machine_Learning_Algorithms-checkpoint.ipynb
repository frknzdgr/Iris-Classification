> I applied six different machine learning classification algorithms to the Iris dataset, visualized the confusion matrix for each model, and compared their performance metrics at the end.

> The results showed that **Random Forest Classifier** achieved the perfect classification performance. Logistic Regression and Support Vector Classifier followed closely, while the remaining models also performed strongly with only minor differences.

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Methods

```python
def confusion_matrix_heatmap(algorithm, y_test, y_pred):
    
    cm = confusion_matrix(y_test, y_pred)
  
    labels = sorted(list(set(y_test)))

    cm_df = pd.DataFrame(cm, index=labels, columns=labels)

    plt.figure(figsize=(6,4))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap="Blues")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{algorithm} Confusion Matrix Heatmap")
    plt.show()

    print("\n")


def results_df(y_test, y_pred):
    results = pd.DataFrame({
    "y_test": y_test.reset_index(drop=True),
    "y_pred": y_pred
    })

    return results

    print("\n\n")



def failed_row(df):
    wrong = df[df["y_test"] != df["y_pred"]]
    if len(wrong) > 0:
        return wrong
    else:
        print("All Predictions are True")

    

def metrics_df(y_test, y_pred, algorithm_name="Model"):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    df = pd.DataFrame({
        "Model": [algorithm_name],
        "Accuracy": [accuracy],
        "Precision": [precision],
        "Recall": [recall],
        "F1-Score": [f1]
    })


    return df
    print("\n\n")
```

## Data and Data Preprocessing

```python
df=pd.read_excel("Iris.xlsx")
df
```

```python
df["iris"]=df["iris"].replace("Iris-setosa","Setosa")
df["iris"]=df["iris"].replace("Iris-versicolor","Versicolor")
df["iris"]=df["iris"].replace("Iris-virginica","Virginica")
df
```

```python
df.iris.unique()
```

```python
sns.countplot(x=df["iris"])
```

```python
sns.scatterplot(
    data=df, x="petal length", y="petal width", hue="iris"
)
```

```python
x=df.iloc[:,:4] # independent
y=df["iris"] # dependent
```

```python
print(x,"\n\n",y)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.33, random_state=42)

print("X_TRAIN\n\n",x_train,"\n\n\n\nY_TRAIN\n\n",y_train,"\n\n\n\nX_TEST\n\n",x_test,"\n\n\n\nY_TEST\n\n",y_test)
```

### Scaling

```python
from sklearn import preprocessing

sc=preprocessing.StandardScaler()

X_train= sc.fit_transform(x_train)
X_test=sc.transform(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train,y_train)

y_pred_logreg = logr.predict(X_test)

display(results_df(y_test, y_pred_logreg))
```

```python
display(failed_row(results_df(y_test, y_pred_logreg)))
confusion_matrix_heatmap("Logistic Regression",y_test,y_pred_logreg)

logreg_metrics = metrics_df(y_test, y_pred_logreg, "Logistic Regression")
logreg_metrics
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1, metric='minkowski')
knn.fit(X_train,y_train)

y_pred_knn = knn.predict(X_test)

display(results_df(y_test, y_pred_knn))
```

```python
display(failed_row(results_df(y_test, y_pred_knn)))
confusion_matrix_heatmap("KNN", y_test, y_pred_knn)

knn_metrics = metrics_df(y_test, y_pred_knn,"KNN")
knn_metrics
```

## SVC

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train,y_train)

y_pred_svc = svc.predict(X_test)

display(results_df(y_test, y_pred_svc))
```

```python
display(failed_row(results_df(y_test, y_pred_svc)))
confusion_matrix_heatmap("SVC",y_test,y_pred_svc)

svc_metrics = metrics_df(y_test, y_pred_svc, "SVC")
svc_metrics
```

## Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred_gnb = gnb.predict(X_test)

display(results_df(y_test, y_pred_gnb))
```

```python
display(failed_row(results_df(y_test, y_pred_gnb)))
confusion_matrix_heatmap("GNB", y_test, y_pred_gnb)

gnb_metrics = metrics_df(y_test, y_pred_gnb, "GNB")
gnb_metrics
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy')

dtc.fit(X_train,y_train)
y_pred_dt = dtc.predict(X_test)

display(results_df(y_test, y_pred_dt))
```

```python
display(failed_row(results_df(y_test, y_pred_gnb)))
confusion_matrix_heatmap("Decision Tree Classifier", y_test, y_pred_gnb)

dt_metrics = metrics_df(y_test, y_pred_gnb, "Decision Tree Classifier")
dt_metrics
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=5, criterion = 'entropy')
rfc.fit(X_train,y_train)

y_pred_rf = rfc.predict(X_test)

display(results_df(y_test, y_pred_rf))
```

```python
display(failed_row(results_df(y_test, y_pred_rf)))
confusion_matrix_heatmap("Random Forest Classifier", y_test, y_pred_rf)

rf_metrics = metrics_df(y_test, y_pred_rf, "Random Forest Classifier")
rf_metrics
```

## Evaluation of Models with Performance Metrics

> I ranked the models by **F1-Score**  because the iris dataset is a **multi-class classification problem**, and F1 provides a more reliable overall performance measure than accuracy.

```python
metrics_all = pd.concat([logreg_metrics, knn_metrics, svc_metrics, gnb_metrics, dt_metrics, rf_metrics], ignore_index=True)
metrics_all = metrics_all.sort_values(by = "F1-Score", ascending=False).reset_index(drop = True)
metrics_all
```
